        	R[i,j]~dbern(pi[i,j])

                               +b[6]*y[i,5]+b[7]*y[i,6]+b[8]*y[i,7]+b[9]*y[i,8]
     	}
	     # measurement models
     	for(j in 1:P){ 
			y[i,j]~dnorm(mu[i,j],psi[j]) 
			ephat[i,j]<-y[i,j]-mu[i,j]
		}
        mu[i,1]<- vu[1]+xxi[i]
        mu[i,2]<- vu[2]+lam[1]*xxi[i]
        mu[i,3]<- vu[3]+xi[i,1]
